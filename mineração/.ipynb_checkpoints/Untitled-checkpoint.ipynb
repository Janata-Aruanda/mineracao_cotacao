Coletei dados do yahoo finanças = https://br.financas.yahoo.com/



```python
#Baixando as Bibliotecas 

from pandas_datareader import data as web

import pandas as pd 

import matplotlib.pyplot as plt
```

```python
#COLETANDO DADOS DA BOSVEPA PELO WERB (YOUTUBE)
cotacao_bosvepa = web.DataReader('^BVSP' ,  data_source='yahoo',  start="10/01/2022", end="31/10/2022")
display(cotacao_bosvepa)
```

```python
#RENOMEANDO NOME DAS COLUNAS 
cotacao_bosvepa.rename(columns={"High" : "Alto" , "Low" : "Baixo" , "Open" : "Aberto" , "Close" : "Pero" , "Adj Close" : "Adj Fechamento"},inplace=True)

```

```python
#Incluindndo index na tabela BOSVEPA
cotacao_bosvepa.reset_index(inplace=True)

```

```python
#Visualizando tabela
display(cotacao_bosvepa)

#Criando grafico com valor da coluna
cotacao_bosvepa['Adj Fechamento'].plot(figsize=(15 , 8))

#Exibindo grafco 
plt.show()
```

```python
#COLETANDO DADOS DA PETROBRAS PELO WERB (YOUTUBE)
cotacao_petrobras = web.DataReader('PETR4.SA' ,  data_source='yahoo',  start="10/01/2022", end="31/10/2022")
display(cotacao_petrobras)
```

```python
#RENOMEANDO COLUNAS DA TABELA PETROBRAS
cotacao_petrobras.rename(columns={"High" : "Alto" , "Low" : "Baixo" , "Open" : "Aberto" , "Close" : "Pero" , "Adj Close" : "Adj Fechamento"},inplace=True)
```

```python
#Incluindndo index na tabela PETROBAS
cotacao_petrobras.reset_index(inplace=True)
```

```python
#Visualizando tabela
display(cotacao_petrobras)

#Criando grafico com valor da coluna
cotacao_petrobras['Adj Fechamento'].plot(figsize=(15 , 10))

#Exibindo grafco 
plt.show()
```
